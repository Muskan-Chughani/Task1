```python
from google.colab import drive
import pandas as pd
import numpy as np
from scipy import stats


drive.mount('/content/drive')


file_path = '/content/drive/MyDrive/internship/critic_reviews.csv'
df = pd.read_csv(file_path)

print(df.head())
print(df.info())
print(df.describe())


print(df.isnull().sum())
df = df.dropna()


print(df.duplicated().sum())
df = df.drop_duplicates()


numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
df = df[(np.abs(stats.zscore(df[numerical_cols])) < 3).all(axis=1)]


cleaned_file_path = '/content/drive/MyDrive/internship/cleaned_critic_reviews.csv'
df.to_csv(cleaned_file_path, index=False)
```

```python
from google.colab import drive
import pandas as pd
import os


drive.mount('/content/drive')


file_path = '/content/drive/MyDrive/internship/online_course_engagement_data.csv'
print(os.listdir('/content/drive/MyDrive/internship/'))


with open(file_path, 'r') as file:
    for i in range(10):
        print(file.readline())


try:
    df = pd.read_csv(file_path)
    if df.empty:
        raise ValueError("The file is empty.")
    print(df.head())
except pd.errors.EmptyDataError:
    print("No columns to parse from file")
except ValueError as ve:
    print(ve)
except Exception as e:
    print(f"An error occurred: {e}")


if 'df' in locals():

    print(df.info())
    print(df.describe())


    print(df.isnull().sum())
    df = df.dropna()


    print(df.duplicated().sum())
    df = df.drop_duplicates()



    from scipy import stats
    numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
    df = df[(np.abs(stats.zscore(df[numerical_cols])) < 3).all(axis=1)]




    cleaned_file_path = '/content/drive/MyDrive/internship/cleaned_online_course_engagement_data.csv'
    df.to_csv(cleaned_file_path, index=False)
    print("Data cleaned and saved successfully.")
else:
    print("Data cleaning skipped as the dataset could not be loaded.")
```
